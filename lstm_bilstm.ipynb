```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow import keras
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import KFold
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc

# Load the dataset
df = pd.read_csv('R2_ChatGPt_dataset.csv', encoding='latin1')

# Preprocess the text data
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    tokens = text.split()
    filtered_tokens = [token for token in tokens if token not in stop_words]
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    return ' '.join(lemmatized_tokens)

df['cleaned_text'] = df['Base_Reviews'].apply(clean_text)
X = df['cleaned_text'].values
y = df['category'].map({'anger': 0, 'confusion': 1, 'disappointment': 2, 'distrust': 3,
                        'disgust': 4, 'frustration': 5, 'fear': 6, 'sadness': 7}).values

# Convert the labels to categorical variables
y = keras.utils.to_categorical(y)

# Tokenize the data
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(num_words=2000)
tokenizer.fit_on_texts(X)
X = tokenizer.texts_to_sequences(X)
X = pad_sequences(X, maxlen=100)

# Apply oversampling to balance the classes
oversampler = RandomOverSampler(random_state=42)
X_resampled, y_resampled = oversampler.fit_resample(X, y)

# Define the LSTM model
lstm_model = keras.models.Sequential([
    keras.layers.Embedding(2000, 100, input_length=100),
    keras.layers.LSTM(128, dropout=0.2, recurrent_dropout=0.2),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(len(y[0]), activation='softmax')
])
lstm_model.compile(loss='categorical_crossentropy', optimizer=keras.optimizers.Adam(learning_rate=0.001), metrics=['accuracy'])

# Define the BiLSTM model
bilstm_model = keras.models.Sequential([
    keras.layers.Embedding(2000, 100, input_length=100),
    keras.layers.Bidirectional(keras.layers.LSTM(128, dropout=0.2, recurrent_dropout=0.2)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(len(y[0]), activation='softmax')
])
bilstm_model.compile(loss='categorical_crossentropy', optimizer=keras.optimizers.Adam(learning_rate=0.001), metrics=['accuracy'])

# Define model training with callbacks for early stopping and accuracy tracking
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

def train_model(model, X_train, y_train, X_val, y_val, model_name):
    early_stopping = EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)
    history = model.fit(X_train, y_train, epochs=10, batch_size=128, verbose=1,
                        validation_data=(X_val, y_val), callbacks=[early_stopping])
    # Save model
    model.save(f"{model_name}_best_model.h5")
    return history

# Perform k-fold cross-validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)
lstm_acc_scores = []
bilstm_acc_scores = []
lstm_histories = []
bilstm_histories = []
all_y_true = []
all_lstm_pred = []
all_bilstm_pred = []
lstm_probs_all = []
bilstm_probs_all = []
y_test_all = []

for train_index, test_index in kf.split(X_resampled):
    X_train, X_test = X_resampled[train_index], X_resampled[test_index]
    y_train, y_test = y_resampled[train_index], y_resampled[test_index]

    lstm_history = train_model(lstm_model, X_train, y_train, X_test, y_test, "LSTM")
    lstm_histories.append(lstm_history.history)
    lstm_acc_scores.append(lstm_model.evaluate(X_test, y_test, verbose=0)[1])
    lstm_pred = np.argmax(lstm_model.predict(X_test), axis=-1)
    lstm_probs = lstm_model.predict(X_test)

    bilstm_history = train_model(bilstm_model, X_train, y_train, X_test, y_test, "BiLSTM")
    bilstm_histories.append(bilstm_history.history)
    bilstm_acc_scores.append(bilstm_model.evaluate(X_test, y_test, verbose=0)[1])
    bilstm_pred = np.argmax(bilstm_model.predict(X_test), axis=-1)
    bilstm_probs = bilstm_model.predict(X_test)

    y_true = np.argmax(y_test, axis=-1)
    all_y_true.extend(y_true)
    all_lstm_pred.extend(lstm_pred)
    all_bilstm_pred.extend(bilstm_pred)
    lstm_probs_all.extend(lstm_probs)
    bilstm_probs_all.extend(bilstm_probs)
    y_test_all.extend(y_test)

# Calculate the average accuracy
avg_lstm_acc = np.mean(lstm_acc_scores)
avg_bilstm_acc = np.mean(bilstm_acc_scores)
print(f"LSTM: Average Accuracy = {avg_lstm_acc:.2f}")
print(f"BiLSTM: Average Accuracy = {avg_bilstm_acc:.2f}")

# Confusion matrices
lstm_cm = confusion_matrix(all_y_true, all_lstm_pred)
bilstm_cm = confusion_matrix(all_y_true, all_bilstm_pred)

# Plot confusion matrices
plt.figure(figsize=(8, 6))
sns.heatmap(lstm_cm, annot=True, fmt='d', cmap='Blues', cbar_kws={'label': 'Scale'})
plt.title('LSTM Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig('LSTM_Confusion_Matrix.png', dpi=300)  # Save the plot as a PNG file with 300 DPI
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(bilstm_cm, annot=True, fmt='d', cmap='Blues', cbar_kws={'label': 'Scale'})
plt.title('BiLSTM Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig('BiLSTM_Confusion_Matrix.png', dpi=300)  # Save the plot as a PNG file with 300 DPI
plt.show()

# Classification reports
print("LSTM Classification Report:")
print(classification_report(all_y_true, all_lstm_pred))
print("BiLSTM Classification Report:")
print(classification_report(all_y_true, all_bilstm_pred))

# ROC Curves and AUC Calculation
n_classes = y_test_all[0].shape[0]
lstm_probs_all = np.array(lstm_probs_all)
bilstm_probs_all = np.array(bilstm_probs_all)
y_test_all = np.array(y_test_all)

# Compute ROC curve and ROC area for each class
fpr_lstm, tpr_lstm, roc_auc_lstm = {}, {}, {}
fpr_bilstm, tpr_bilstm, roc_auc_bilstm = {}, {}, {}

for i in range(n_classes):
    fpr_lstm[i], tpr_lstm[i], _ = roc_curve(y_test_all[:, i], lstm_probs_all[:, i])
    roc_auc_lstm[i] = auc(fpr_lstm[i], tpr_lstm[i])
    fpr_bilstm[i], tpr_bilstm[i], _ = roc_curve(y_test_all[:, i], bilstm_probs_all[:, i])
    roc_auc_bilstm[i] = auc(fpr_bilstm[i], tpr_bilstm[i])

# Plotting ROC Curves for LSTM
plt.figure(figsize=(10, 8))
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'pink']
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr_lstm[i], tpr_lstm[i], color=color, lw=2,
             label='Class {0} (area = {1:0.2f})'.format(i, roc_auc_lstm[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for LSTM')
plt.legend(loc="lower right")
plt.savefig('LSTM_ROC_Curve.png', dpi=300)  # Save the plot as a PNG file with 300 DPI
plt.show()

# Plotting ROC Curves for BiLSTM
plt.figure(figsize=(10, 8))
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr_bilstm[i], tpr_bilstm[i], color=color, lw=2,
             label='Class {0} (area = {1:0.2f})'.format(i, roc_auc_bilstm[i]))

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for BiLSTM')
plt.legend(loc="lower right")
plt.savefig('BiLSTM_ROC_Curve.png', dpi=300)  # Save the plot as a PNG file with 300 DPI
plt.show()

```

```python
# Function to plot average training and validation accuracy and loss over all folds
def plot_avg_accuracy_loss(histories_lstm, histories_bilstm):
    # Determine the shortest training duration among all folds
    min_epochs_lstm = min(len(history['accuracy']) for history in histories_lstm)
    min_epochs_bilstm = min(len(history['accuracy']) for history in histories_bilstm)
    min_epochs = min(min_epochs_lstm, min_epochs_bilstm)
    
    # Collect data only up to the shortest training duration
    lstm_acc = [history['accuracy'][:min_epochs] for history in histories_lstm]
    lstm_val_acc = [history['val_accuracy'][:min_epochs] for history in histories_lstm]
    lstm_loss = [history['loss'][:min_epochs] for history in histories_lstm]
    lstm_val_loss = [history['val_loss'][:min_epochs] for history in histories_lstm]
    bilstm_acc = [history['accuracy'][:min_epochs] for history in histories_bilstm]
    bilstm_val_acc = [history['val_accuracy'][:min_epochs] for history in histories_bilstm]
    bilstm_loss = [history['loss'][:min_epochs] for history in histories_bilstm]
    bilstm_val_loss = [history['val_loss'][:min_epochs] for history in histories_bilstm]

    # Average the data across folds
    avg_lstm_acc = np.mean(lstm_acc, axis=0)
    avg_lstm_val_acc = np.mean(lstm_val_acc, axis=0)
    avg_lstm_loss = np.mean(lstm_loss, axis=0)
    avg_lstm_val_loss = np.mean(lstm_val_loss, axis=0)
    avg_bilstm_acc = np.mean(bilstm_acc, axis=0)
    avg_bilstm_val_acc = np.mean(bilstm_val_acc, axis=0)
    avg_bilstm_loss = np.mean(bilstm_loss, axis=0)
    avg_bilstm_val_loss = np.mean(bilstm_val_loss, axis=0)

    plt.figure(figsize=(12, 10))
    plt.subplot(2, 1, 1)
    plt.plot(avg_lstm_acc, 'b--', label='LSTM Training Accuracy')
    plt.plot(avg_lstm_val_acc, 'r-', label='LSTM Validation Accuracy')
    plt.plot(avg_bilstm_acc, 'g--', label='BiLSTM Training Accuracy')
    plt.plot(avg_bilstm_val_acc, 'm-', label='BiLSTM Validation Accuracy')
    plt.title('(a) Average Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend(loc='best')

    plt.subplot(2, 1, 2)
    plt.plot(avg_lstm_loss, 'b--', label='LSTM Training Loss')
    plt.plot(avg_lstm_val_loss, 'r-', label='LSTM Validation Loss')
    plt.plot(avg_bilstm_loss, 'g--', label='BiLSTM Training Loss')
    plt.plot(avg_bilstm_val_loss, 'm-', label='BiLSTM Validation Loss')
    plt.title('(b) Average Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend(loc='best')

    plt.tight_layout()
    plt.savefig('Avg_Training_Validation_Accuracy_Loss.png', dpi=300)  # Save the plot as a PNG file with 300 DPI
    plt.show()

# Plot average training and validation accuracy and loss
plot_avg_accuracy_loss(lstm_histories, bilstm_histories)
```
